#**Food Consumption Analysis**

```python
# Import relevant libraries

import pandas as pd
import numpy as np
import plotly.express as pt
import matplotlib.pyplot as plt
```

```python
# import data set from source to perform analysis

data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data Sets/Indian_Food_Dataset.csv")

# First 5 Rows in the dataset

data.head()
```

**Data Wrangling / Data Cleaning**

```python
# See all the coloumn headers

data.columns
```

```python
# See all the info about the imported data

data.info()
```

```python
# Check whether there is any NULL value in the columns or not

data.isnull().any()
```

```python
# Check the number of NULL values column wise in the dataset

data.isnull().sum()
```

```python
# Clean the data set by replacing the null values

data=data.replace(-1,np.nan)
data=data.replace("-1",np.nan)
```

```python
data.isnull().sum()
```

```python
data.shape
```

**Creating useful and readable insights**

```python
veg_nveg=data.diet.value_counts().reset_index()
veg_nveg
```

```python
veg_nveg.columns=['diet','count']
fig=pt.pie(veg_nveg, values='count', names='diet', title= "Veg and Non-veg ratio", color_discrete_sequence=['green','red'])
fig.show()
```

```python
cooking_time=data[['cook_time','name']]
```

```python
cooking_time.head()
```

```python
cooking_time=cooking_time.sort_values(['cook_time'],ascending=True)
ten_cooking_time=cooking_time.head(10)
ten_cooking_time
```

```python
figs=pt.bar(ten_cooking_time,y='cook_time',x='name',title='dishes based on cooking time',color_discrete_sequence=['green'] )
figs.show()
```
